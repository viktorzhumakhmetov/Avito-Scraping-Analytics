### Веб парсинг крупнейшего российского сайта объявлений avito.ru.
В данной работе будет проведен сбор информации из объявлений о продаже автомобилей немецких брендов на 29.05.23.
Сбор информации проводится с помощью библиотек Selenium, Requests и BS4.
Все собранные данные будут преобразованы в Pandas DataFrame и на основании этой информации проводится аналитическое исследование.

```python
import requests
from bs4 import BeautifulSoup
from selenium.webdriver import Chrome
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
from tqdm import tqdm
from time import sleep
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
data = []
browser = Chrome('/Users/viktorzumahmetov/Desktop/chromedriver')
```

```python
# BMW
for p in tqdm(range(1,101)):
    url = f'https://www.avito.ru/all/avtomobili/bmw-ASgBAgICAUTgtg3klyg?cd=1&p={p}'
    browser.get(url)
    sleep(1)
    soup = BeautifulSoup(browser.page_source, 'html')
    cars = soup.findAll('div', class_='iva-item-root-_lk9K')
    for car in cars:
        try:
            brand = car.find('h3','title-root-zZCwT').text.split(', ')[0].split()[0]
            model = ' '.join(car.find('h3','title-root-zZCwT').text.split(', ')[0].split()[1:])
            year = int(car.find('h3','title-root-zZCwT').text.split(', ')[1])
            mileage = int(car.find('div', class_='iva-item-text-Ge6dR').text.replace('\xa0',' ').split(', ')[0].replace('км','').replace(' ',''))
            fuel_type = car.find('div', class_='iva-item-text-Ge6dR').text.replace('\xa0',' ').split(', ')[-1]
            wheel_drive = car.find('div', class_='iva-item-text-Ge6dR').text.replace('\xa0',' ').split(', ')[-2]
            vehicle_type = car.find('div', class_='iva-item-text-Ge6dR').text.replace('\xa0',' ').split(', ')[-3]
            try:
                engine_vol = float(car.find('div', class_='iva-item-text-Ge6dR').text.replace('\xa0',' ').split()[3])
            except:
                engine_vol = 0
            if engine_vol == 0:
                transmission = 'AT'
            else:
                transmission = car.find('div', class_='iva-item-text-Ge6dR').text.replace('\xa0',' ').split(', ')[1].split()[1]
            horsepower = int(car.find('div', class_='iva-item-text-Ge6dR').text.replace('\xa0',' ').split(', ')[1].split()[-2][1:])
            price = int(car.find('span', class_ = 'price-text-_YGDY').text.replace('\xa0','')[:-1])
            region = car.find('div',class_='geo-address-fhHd0').text
            data.append([brand, model, year, mileage, fuel_type, wheel_drive,vehicle_type,engine_vol,transmission,horsepower,price,region])
        except:
            pass
```

```python
# Выделение названия области из строки в столбце региона через применение функции

def get_region(s):
    if ' — ' in s:
        return s.split(' — ')[0]
    elif ',' in s:
        return s.split(', ')[0]
    else:
        return s

df.region = df.region.apply(get_region)
```

```python
# Округление пробега до тысяч километров

df.mileage = df.mileage.apply(lambda x: x//1000)
```

## Основные выводы по продаваемым в России автомобилям немецких производителей:
* Больше всего в продаже автомобилей 2012 и 2019 года, возможно эти годы были особенно успешными по продажам новых авто.
* Основная часть автомобилей продается с пробегом от 90 до 160 тысяч км.
* Наиболее распространены двигатели объемом 2.0 и 3.0 литра.
* Среди двух литровых двигателей наиболее популярен бензиновый вариант, среди трехлитровых - дизельный.
* Основная масса автомобилей по мощности сосредоточена в диапазоне 150-200 л.с.
* Самый популярный тип трансмиссии - автоматическая, среди этой категории чаще всего встречаются внедорожники и седаны.
* Наименее популярный тип трансмиссии - вариатор, который кроме седанов используется крайне редко.
* Наиболее мощные автомобили в большинстве случаев оснащены роботизированной трансмиссией и задним приводом.
* Вариатор встречается только в сочетании с передним приводом.
* По типу привода самый многочисленный класс - полноприводный, в котором в основном преобладают внедорожники, в остальных случаях наиболее часто встречаются седаны.
* Тенденция изменения среднего объема двигателя в последние годы идет на спад в связи с широким распространением турбонаддува и электромобилей, но при этом с 2018 наблюдается стабильный и внушительный рост мощности, однако с 2000 по 2010 годы объемы увеличивались, при этом так же увеличивалась мощность.
* В среднем переднеприводный класс автомобилей обходится дешевле и менее подвержен сильно выраженной потере стоимости.
